```python
# Verilen string'i ters çevirmek
def terscevir(s):
    if (type(s) != str):
        raise ValueError("Lütfen doğru bir input girin.")
    else:
        return s[::-1]
    
```

```python
print(terscevir(123
                ))
```

```python
print(terscevir("sukru"))
```

```python
try:
    print(terscevir("sukru"))
except ValueError:
    print("Fonksiyonda bir hata oluştu.")
```
