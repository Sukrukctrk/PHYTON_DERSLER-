```python
#  hesap makinesi kodu 

a = int(input("bir a değeri giriniz"))
b = int(input("bir b değeri giriniz"))


işlem = input( "yaptırmak istediniz işlemi giriniz")

if işlem== "+":
    print(a,"+",b,"=",a +b)

elif  işlem == "-":
    print(a,"-",b,"=",a-b)

elif işlem== "*":
    print(a,"*",b,"=",a*b)


else :
  işlem == "/"
  print(a,"/",b,"=",a/b)
```
