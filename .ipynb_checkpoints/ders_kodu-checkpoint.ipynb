```python
#  asal sayı bulma programı

def asal_mı(x):
  if x==1:
     return False
        
  elif x==2:
            return True
        
  else:
       for i in range(2,x):
            if x%i == 0:
                 return False
            
            else :
                 return True
       
```

```python
asal_mı(12)
```

```python
#  bir sayının tam bölenlerini bulma

def tam_bölen (x):
    tam_bölenler =[]
    
 
    for i in range(2,x):
        if  x%i==0:
              
         tam_bölenler.append(i)


        return tam_bölen
        

    
```

```python
tam_bölen(12)
```
