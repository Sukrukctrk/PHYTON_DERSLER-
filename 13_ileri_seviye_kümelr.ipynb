```python
# Eleman Eklemek : add() metodu
# Kümeye eleman eklemimizi sağlar. Aynı eleman eklenmeye
# çalışırsa hata vermez ve herhangi bir ekleme işlemi yapmaz.


x = {1,2,3}
x.add(4)
x
```

```python
# İki kümenin farkı : difference() metodu
# Bu metod birinci kümenin ikinci kümeden farkını döner.

# küme1.difference(küme2)  Küme1'in Küme2'den farkı
```

```python
küme1 = {1,2,3,10,34,100,-2}
küme2 = {1,2,23,34,-1}
küme1.difference(küme2)
```

```python
küme2.difference(küme1)
```

```python
# Kümeden Eleman Çıkartmak : discard() metodu
# İçine verilen değeri kümeden çıkartır. Eğer kümede öyle bir değer yoksa,
# bu metod hiçbir şey yapmaz(Hata vermez).

küme1.discard(3)
```

```python
küme1
```

```python
#  intersection metodu iki kümenin kesişimini bulmamızı saglar

küme1.intersection(küme2)
```

```python

küme1 = {1,2,3,10,34,100,-2}
küme2 = {1,2,23,34,-1}


küme1.isdisjoint(küme2)
#   isdisjoint ayrık kümemi demek
```

```python
küme1 ={1,2,3}
küme2 ={1,2,3,4,5}

küme1.issubset(küme2)
#  issubset alt kumesimi demek
```

```python
küme1.union(küme2) 
#  union birleşim demek
```
