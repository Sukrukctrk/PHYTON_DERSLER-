```python
def diziyi_yaz(n):
    """
    1 den n ye kadar olan sayıları yazdırır
    
    """
    for i in range(n):
        print(i , end= " ")


diziyi_yaz(10)
```

```python
def diziyi_döndürür(n):

    """  
    1"den n" ye kadar olan sayıları string olan döndüren fonksiyon
    diziyi_döndürür(3):> " 1 2 3 "
     diziyi_döndürür(5) => " 1 2 3 4 5 "
    
    """
   
    a = ""
    for i in range(n):
       
       a = a + str(i) + " "

    return a

      

```

```python
sayilar = diziyi_döndürür(5)
print(sayilar)
print(sayilar[3:])
```

```python
#  n! sayısını hesaplayan bir fonksiyon

def faktoryel(n):
    sonuç = 1

    for i in range(1, n+1):
        sonuç = sonuç * i

    return sonuç

print(faktoryel(6))
```

```python
#  bir sayının asal sayı olup olamdığını bulan fonksiyon

def asal_mı(x):
   
    """   
    eğer asalsa true değilse false dön
    
    """

    for i in range(2, x):
        if  x % i == 0:

            return False
        
    return True


print(asal_mı( 2 ), asal_mı(900), asal_mı(41))
```

```python
def tüm_asallar(n):

    for i in range(2, n):
        if asal_mı(i):
            print(i , end= ", ")

tüm_asallar(90)
        
    
```

```python
def asallar_listesi_oluştur(n):

    sonuc =[]

    for i in range(2, n+1):
        if asal_mı(i):
            sonuc.append(i)

    return sonuc


my_list = asallar_listesi_oluştur(51)
print(my_list)
```

```python
def çift_mi(n):

    kalan= n % 2 
    karar = False

    if kalan == 0:
        karar = True

    return karar


print(çift_mi(4), çift_mi(5) )


```

```python
def tek_mi(n):
    x= çift_mi(n)

    karar = not x
    return karar


print(tek_mi(3), tek_mi(120))
```

```python
#  verrilen listedeki en büyük sayıyı bulan fonksiyon

def en_buyuk_sayı(l):
    en_buyuk= l[0]
    for el in l :
        if el> en_buyuk:
           en_buyuk= el


    return en_buyuk

print(en_buyuk_sayı([1,2,3,4,5]))
```

```python
# bir stringin palindrom olup olmadığını bulan program yapınız
#  ata , abcba birer palindromdur

def palindrom_mu(s):

    uzunluk =len(s)
    for i in range(int(uzunluk / 2)):
        if s[i] != [(i+1)* -1]:
            return False
        

    return True

print(palindrom_mu("abcdadh"),palindrom_mu("nalan"),palindrom_mu("ata"))

```
