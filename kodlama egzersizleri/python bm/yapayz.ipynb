```python
bakiye = 1000

while True:
    print("1. Bakiye sorgula")
    print("2. Para yatır")
    print("3. Para çek")
    print("4. Çıkış")

    secim = input("İşlem seçiniz: ")

    if secim == "1":
        print(f"Bakiyeniz: {bakiye} TL")
    elif secim == "2":
        miktar = int(input("Yatırmak istediğiniz tutar: "))
        bakiye += miktar
        print(f"Yeni bakiyeniz: {bakiye} TL")
    elif secim == "3":
        miktar = int(input("Çekmek istediğiniz tutar: "))
        if miktar > bakiye:
            print("Bakiyeniz yetersiz!")
        else:
            bakiye -= miktar
            print(f"Yeni bakiyeniz: {bakiye} TL")
    elif secim == "4":
        print("İyi günler!")
        break
    else:
        print("Geçersiz işlem!")

        
```

```python
while True:
    num = int(input("Lütfen bir sayı girin: "))
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                print(f"{num} asal sayı değildir.")
                break
        else:
            print(f"{num} asal sayıdır.")
    else:
        print(f"{num} asal sayı değildir.")
```

```python
def is_prime(num):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                return False
        else:
            return True
    else:
        return False

while True:
    num = int(input("Lütfen bir sayı girin: "))
    if is_prime(num):
        print(f"{num} asal sayıdır.")
    else:
        print(f"{num} asal sayı değildir.")
```
