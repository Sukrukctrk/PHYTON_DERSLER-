```python
"""
fibonacci serisi
1, 1, 2, 3, 5, 8...

ilk n tane fibonacci serisis elamanını hesaplayalım

"""

t_2 = 1
t_1= 1
N=7

for i in range(2,N):

    simdiki = t_1+ t_2

    t_2 = t_1
    t_1= simdiki

    print(simdiki ,end=" ")




```

```python

# fiboccani sayı dizisi
t_2 = 1
t_1= 1
fib= [t_2, t_1]
N= 100

for i in range(2,N):

    simdiki = t_1+ t_2
    fib.append(simdiki)
    t_2 = t_1
    t_1= simdiki

print(fib)

```
