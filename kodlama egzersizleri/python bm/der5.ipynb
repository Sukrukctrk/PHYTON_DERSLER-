```python
#  [::-1] ifadeyi tersten yazmayı sağlar
#  len kaç karakter oldunu tanımlamamıza yarar


ad = "sukru"

len(ad)
print(ad[::-1])
```

```python
#  upper yazdımız komut içindeki tüm karekterleri büyük yazamamızısağlar
#  lower yazdımız komut içindeki tüm karekterleri küçük harfle yazar
#  title  cümledeki karakterlerin baş harfini büyük yazmamızı sağlar
#  capitalize cümlenin baaş harfini büyük yazmamızı asğlar


ad = ad.upper()
print(ad)
```

```python
ad = ad.capitalize()
print(ad)
```

```python
# strip()ifadesi baştan ve sondan karakter silmemizi sağlar

# split ifadeside boşluk karakterleinden itibaren ayırır

ad= "sukru" +"kocaturk"

ad= ad.strip()
print(ad)
```
