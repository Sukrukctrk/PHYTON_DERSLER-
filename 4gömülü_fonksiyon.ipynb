```python
def asal_mı(x):
    i=2
    if x==1:
        return False
    
    elif x==2:
        return True
    
    else:
        while i<x:

            if x%i==0:
                return False
            
            else:
                return True
       
```

```python
print(asal_mı(13))
```

```python
list(filter(asal_mı,range(1,100)))
```

```python
# zip fonksiyonu kolay yönden listelememize yarar
sonuç= list()
liste1 =[1,2,3,4]
liste2= [5,6,7,8,9]

i=0

while (i<len(liste1) and i<len(liste2)):
    sonuç.append((liste1[i],liste2[i]))
    i+=1
print(sonuç)

```

```python
list(zip(liste1,liste2))
```

```python
#  enumurete fonksiyonu her elamanı indeksleyerek guruplamamızı sağlar

liste1 = ["elma","armut","muz","kiraz"]

sonuç= list()
i= 0

while i< len(liste1):

    sonuç.append((i,liste1[i]))

    i+=1

print(sonuç)

```

```python
list(enumerate(liste1))
```

```python
for i,j in enumerate(liste1):
    print(i,j)
```

```python
def hepsi(liste):
    for i in liste:
        if not i:
            return False
    return True
# Bütün değerler True ise True en az birisi False ise False döndürmek istiyoruz.
liste = [True,False,True,False,True]
        
```

```python
hepsi(liste) # En az birisi False
```

```python
def herhangi(liste):
    for i in liste:
        if i:
            return True
    return False
# Herhangi bir değer True ise True, Hepsi False ise False döndürmek istiyoruz.
```

```python
liste = [True,False,True,False,True]
herhangi(liste)
```
