```python
def faktoriyel(sayı):
    

    
    faktoriyel =1
    if sayı==1 or sayı==0:

        print("faktoriyel",faktoriyel)


    else:
        for i in range(sayı):
            i= i+1
            faktoriyel= i*faktoriyel

        print("faktoriyel",faktoriyel)
            

```
